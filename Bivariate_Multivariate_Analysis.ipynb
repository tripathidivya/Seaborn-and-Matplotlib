```python
import pandas as pd
import seaborn as sns
```

Loading 4 types of datasets

```python
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')
```

# Bivariate Analysis

## 1. Scatterplot( Numerical - Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', data= tips)
```

# Multivariate Analysis

```python
sns.scatterplot(x='total_bill', y='tip', data= tips, hue='sex')
```

```python
sns.scatterplot(x='total_bill', y='tip', data= tips, hue='sex', style ='smoker', size='size' )
```

## 2. Bar Plot( Numerical - Categorical)

```python
print(titanic.head())
```

```python
sns.barplot(x= 'pclass', y='age', data= titanic)
```

```python
sns.barplot(x= 'pclass', y='fare', data= titanic)
```

```python
sns.barplot(x= 'pclass', y='fare', data= titanic, hue='sex')
```

```python
sns.barplot(x= 'pclass', y='age', data= titanic, hue='sex')
```

## 2. Box Plot( Numerical - Categorical)

```python
sns.boxplot(x='age', y = 'sex', data = titanic)
```

```python
sns.boxplot(x='age', y = 'sex', data = titanic, hue='survived')
```

## 2. Distplot ( Numerica - categorical)

```python
print(titanic.head())
```

```python
sns.distplot(titanic[titanic['survived']==0]['age'], hist =False)
sns.distplot(titanic[titanic['survived']==1]['age'], hist =False)
```

## 5. Heatmap( Categorical - Categorical)

```python

pd.crosstab(titanic['pclass'], titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
```

```python
titanic.groupby('pclass')['survived'].mean()
```

```python
(titanic.groupby('pclass')['survived'].mean()*100).plot(kind='bar')
```

```python
titanic.groupby('survived')['pclass'].mean()
```

```python
flights.pivot_table(index='year', columns='month', values='passengers')
```

```python
sns.heatmap(flights.pivot_table(index='year', columns='month', values='passengers'))
```

## 6. ClusterMap ( Categorical - Categorical)

```python
pd.crosstab(titanic['sibsp'], titanic['survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['sibsp'], titanic['survived']))
```

## 8. Line Plot (Numerical - Numerical)

use for time based numbers

```python
flights.head()
```

```python
print(flights['year'].dtype)
```

```python
grouped_sum = flights.groupby('year').sum(numeric_only=True)
print(grouped_sum)
```
