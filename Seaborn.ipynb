# **Scatter Plots***:
1. Use numerical values
2. High level over view of linear relationship
3. Can be used for categorical data but generally used for numerical data
4. use hue, size and style to show plots

Types of functions used:
1. relplot: kind parameter to plot bar or scatter plot
2. scatterplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')  ## toy dataset
```

```python
print(data.head())
```

```python
sns.relplot(x='total_bill', y='tip', data= data)
```

```python
sns.relplot(kind='scatter', x='total_bill', y='tip', data=data, hue='smoker')
```

```python
sns.relplot(kind='scatter', hue='smoker', style='sex', x='total_bill', y='tip', data=data)
```

```python
sns.relplot(kind='scatter', hue='smoker', style='sex', size='size', x='total_bill', y='tip', data=data)
```

```python
sns.relplot(kind='scatter', hue='smoker', style='sex', size='size',sizes=(15, 200), x='total_bill', y='tip', data=data)
```

# Strip plot*

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('fivethirtyeight')
data=sns.load_dataset('tips')
print(data.head())
```

```python
sns.catplot(kind='strip', x='day', y='tip', data= data)
```

```python
sns.catplot(kind='strip', x='day', y='tip', jitter= 0, data= data)
# Jitter isnoise between points
```

# Swarm plot

```python
sns.catplot(kind='swarm', x='day', y='tip', data= data)
#Shows the distribution of the data
# only difference with strip plot is this it gives detailed information of distributions of data
```

```python
sns.catplot(kind='swarm', x='day', y='tip', hue='sex', data= data)
#it does not include size or style options
```

```python
sns.swarmplot(x='day', y='tip',data= data)
# in place of catplot we can directly use swarm plot, but it has some limitation, whereas catplot has more options
```

# Boxplot*

shows data on based of 5 quartile vales minimum, maximum, range

**One numerical data: Univariate analysis**

```python
sns.boxplot(data['tip'])
```

**One numerical data, one categorical data: multivariate analysis**

```python
sns.catplot(x='day', y='tip', kind='box', data= data)
```

```python
sns.catplot(x='day', y='tip', kind='box', hue='sex', data= data)
```

# **Violin Plots**

1. Use to plot numerical data
2. Combination of boxplot with Kernel density plot

```python
sns.violinplot(data['total_bill'])
```

Compare distribution of given variable across some categories

```python
sns.catplot(x='day', y='total_bill', kind='violin', data= data)
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', data= data)
#But this way graph is very conjusted and it is difficult to understand the graph
```

```python
# so to solve this problem we will use below given method for mor categories
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', split=True, data= data)
```

# **Barplot***

numerical Vs Categorical data

```python
sns.catplot(kind='bar', x='smoker', y='total_bill', data=data)
#as C default value is .95 that means that the value for mean population with a confidence of 95% is lies in the given range of black mark
```

```python
sns.catplot(kind='bar', x='smoker', y='total_bill', hue='sex', data=data)
```

```python
sns.catplot(kind='bar', x='smoker', y='total_bill', estimator=np.median, hue='sex', data=data)
```

```python
sns.catplot(kind='bar', x='smoker', y='total_bill', estimator=np.var, hue='sex', data=data)
```

```python
sns.catplot(kind='bar', x='smoker', y='total_bill', estimator=np.std, hue='sex', data=data)
```

# **Heatmaps***

1. use for data visualization
2. Discrete numerical data

```python
flight= sns.load_dataset('flights')
print(flight.head())
```

```python
x=flight.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot = True, fmt='d')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot = True, fmt='d', cmap= 'summer')
```

# **Cluster Map**

1. Basically a heatmap
2. clustered on the basis of similarity
3. similar rows and columns come together

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric='correlation')
```

```python
sns.clustermap(x, metric='correlation', annot= True)
```

```python
sns.clustermap(x, metric='correlation', annot= True, z_score= 0)
```

```python
sns.clustermap(x, metric='correlation', annot= True, row_cluster= False, z_score= 0)
```

# **Joint Plot**

1. basically use for bivariate analysis
2. analysis between two variables
3. work on numerical data
4. show multi panel figure
5. show bi-variate analysis between two variables
6. Show univariate districution of each variable on seperate axis

4 types of plot:
1. Scatter Plot
2. Hexbin Plot
3. Kde Plots
4. Reg Plot

```python
data = sns.load_dataset('tips')
```

```python
print(data)
```

# **Scatter Plot**

```python
sns.jointplot(x='total_bill', y='tip', data=data)
```

# **Hexbin Plot**

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=data)
#hex plot is 2d histogram
#darker: higher values in range
#lighter: less value
```

# **Kde Plots**

1. use for univariate analysis
2. These plots aso known as corn 2 plot
3. it is 2-d probability function
4. Darker shade areas peak coming out of screen

```python
sns.jointplot(x='total_bill', y ='tip', data= data, kind='kde')
```

# **Reg Plot**

1. it is scatter plot between 2 numberical values
2. the best fit line is linear line
3. use for prediction
4. for future value predict future value

```python
sns.jointplot(x='total_bill', y ='tip', data= data, kind='reg')
```

# **Distplots**

1. Use for univariate analysis
2. continous/ Numerical data

Combine 3 things in one graph;
1. Histogram
2. Kernel Density Estimation (PDF)
3. Rugplot

```python
titanic= sns.load_dataset('titanic')
print(titanic.head())
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace = True)
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'], kde=False)
#counting the number of variables in the given bins
```

# **Hist Plot**

```python
sns.distplot(titanic['age'], bins = 10, kde=False)
```

# **Kde Plot**

```python
sns.distplot(titanic['age'], hist=False)
```

# **Rugplot**

```python
sns.distplot(titanic['age'], kde=False, hist= False, rug = True)
```

1. plots the data side by side
2. you can tell if a point is titled on one value or not

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
