```python
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt

draw= norm.ppf(np.random.random(100000), loc=1000, scale=200).astype(int)
sns.distplot(draw)
```

```python
z=[]
for i in draw:
    z.append((i-1000)/200)
    
    
a=np.array(z)

print(a.mean())
print(np.round(a.mean()))
```

```python
print(a.std())
print(np.round(a.std()))
```
