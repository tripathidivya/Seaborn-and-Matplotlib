```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# **Pie Chart**

```python
areas=['Marketing', 'Sales', 'Development', 'HR', 'Customer sUpport']
budget = [2.5, 4, 10, 1, 6]
```

# **Scatter Plot**

```python
x= [23, 45, 67, 87, 12, 22, 43, 44]
y= [45, 11, 1, 67, 78, 99, 12, 90]

plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age Vs Salary Graph')
```

```python
product_id = [1, 2, 3, 4, 5, 6, 7, 8]
online_price= [233, 456, 770, 120, 222, 444, 200, 300]
offline_price = [400, 300, 100, 333, 444, 555, 666, 999]
chor_bazar = [100, 200, 300, 400, 150, 250, 350, 450]

```

```python
plt.plot(product_id, online_price, color= 'red', marker = '+') # by default it is blue
plt.xlabel("Products")
plt.ylabel("Prices")
plt.title("Product Price Comparison")
plt.show()
```

```python
plt.plot(product_id, online_price, color= 'red', marker = 'D', markersize = 10) # by default it is blue
plt.xlabel("Products")
plt.ylabel("Prices")
plt.title("Product Price Comparison")
plt.show()
```

```python
plt.plot(product_id, online_price, color= 'red', marker = 'D', markersize = 10, label='online') # by default it is blue
plt.plot(product_id, offline_price, color= 'yellow', marker = '+', markersize = 10, label='offline')
plt.plot(product_id, chor_bazar, color= 'green', markersize = 10, label='chorbazar')


plt.xlabel("Products")
plt.ylabel("Prices")
plt.title("Product Price Comparison")
plt.legend() #find the blank space and put the label there only
plt.show()
```
